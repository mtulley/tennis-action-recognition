```python
import numpy as np
data = np.load("../dataset/pose_classification_v1.1/train_data.npy")
print("shape:", data.shape)
print("mean:", data.mean())
print("std:", data.std())
print("min:", data.min(), "max:", data.max())
print("nan count:", np.isnan(data).sum())
print("zero frames:", (data == 0).all(axis=(1,2,3,4)).sum())
```
